```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Decision Tree', 'Random Forest', 'XGBoost']
accuracy = [33, 39, 83]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Create bar chart
plt.figure(figsize=(8, 6))
plt.bar(models, accuracy, color=colors, width=0.5)

# Labels and title
plt.xlabel('Prediction Models')
plt.ylabel('Accuracy (%)')
plt.title('Flight Delay Predictions')
plt.ylim(0, 100)

# Annotate values
for i, v in enumerate(accuracy):
    plt.text(i, v + 2, f"{v}%", ha='center', fontsize=12, fontweight='bold')

# Show grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
